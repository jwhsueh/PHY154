```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.optimize import curve_fit

with open("hubble1.dat") as textfile:
    lines = [line.split() for line in textfile]

#print lines

radius1 = np.zeros(24)
velocity1 = np.zeros(24)

for i in range(0, 23):
    radius1[i] = float(lines[i][2])
    velocity1[i] = float(lines[i][3])
    #print radius1[i]    
    
    
with open("hubble2.dat") as textfile:
    lines = [line.split() for line in textfile]

#print lines

radius2 = np.zeros(76)
velocity2 = np.zeros(76)

for i in range(0, 75):
    radius2[i] = float(lines[i][4])
    velocity2[i] = float(lines[i][5])
    #print radius2[i]    
    
    
fig = plt.figure(1, figsize=(10,40))

ax = fig.add_subplot(511)

plt.xlabel("Distance (Mpc)")
plt.ylabel("Velocity (km/s)")

ax.scatter(radius1, velocity1)

def line(x, a):
    return a*x

a, err = curve_fit(line, radius1, velocity1)

perr = np.sqrt(np.diag(err))

r1 = np.linspace(0,2.5,100)

ax.plot(r1, a[0]*r1,label='Best Fit')
ax.plot(r1, (a[0]-perr)*r1,label='Lower Limit Fit')
ax.plot(r1, (a[0]+perr)*r1,label='Upper Limit Fit')
plt.xlim([0,2.5])
legend = plt.legend(loc='upper left')


ax2 = fig.add_subplot(512)

a2, err2 = curve_fit(line, radius2, velocity2)

perr2 = np.sqrt(np.diag(err2))

r2 = np.linspace(0,1000,1000)

ax2.plot(r2, a2[0]*r2, label='Best Fit')
ax2.plot(r2, (a2[0]-1)*r2, label='Lower Limit Fit')
ax2.plot(r2, (a2[0]+1)*r2, label='Upper Limit Fit')
plt.xlim([0,1000])
plt.ylim([0,120000])

ax2.scatter(radius2, velocity2)
plt.xlabel("Distance (Mpc)")
plt.ylabel("Velocity (km/s)")

legend = plt.legend(loc='upper left')
plt.show()


print a[0]
print perr
print a2[0]
print perr2
```

$\textbf{2(a)}$ The Newtonian approach can be a good approximation on the galaxy scale. As long as the distances are much smaller than the radius of curvature of spacetime, the GR approach is not necessary. As the distances become of the order of the radius of curvature of spacetime, the effects of GR become non-trivial and the Newtonian approach is no longer valid.

$\textbf{2(b)}$ Treat the expanding universe as a sphere of radius $a x$ where x is the comoving radius and with mass $M = $ const. The gravitational acceleration is given by 
$$\frac{d^{2}a}{dt^{2}} x = -\frac{G M}{(a x)^{2}}$$

Use $M = \frac{4}{3} \pi \rho (x a)^3$ and divide both sides by x to isolate $\frac{d^{2}a}{dt^{2}}$:

$$\frac{d^{2}a}{dt^{2}} = -\frac{G 4 \pi \rho a}{3}$$

$\textbf{2(c)}$ Multiply both sides by $\frac{da}{dt}$:

$$\frac{d^{2}a}{dt^{2}}  \frac{da}{dt} = -\frac{G 4 \pi \rho a}{3}  \frac{da}{dt}$$

Integrate and include K as a constant of integration on the r.h.s:

$$\frac{1}{2}\frac{da}{dt} = \frac{G 2 \pi \rho a^{2}}{3} + K$$

$$\frac{da}{dt} = \frac{G 4 \pi \rho a^{2}}{3} + K$$

This is now an energy equation instead of a force equation. This equation gives the relationship between the kinetic and potential energy of the system.

$\textbf{2(d)}$ If K > 0, $\frac{da}{dt}$ could be negative or zero. This means the sphere will either have a constant radius or collapse.

If K = 0 or K < 0, $\frac{da}{dt}$ would be positive and the sphere would expand. However, if K = 0 then $\frac{da}{dt}$ goes to 0 as t goes to infinity whereas for K < 0 $\frac{da}{dt}$ will always be positive. 

$\textbf{2(e)}$ $$p_{cr} = \frac{3 H_{0}^{2}}{8 \pi G} = 1.88 \times 10^{-29} h^{2} g cm^{-3}$$ 
is the critical density in the case of K = 0. It is the density found by setting $t = t_{0}$ and $H_{0} = \frac{da}{dt}$ for $t_{0}$. The critical density is the density needed for the universe to expand consistently.

$\textbf{2(f)}$ Matter can be considered pressure-free because the momentum is small compared to $\rho c$. Most of the particles in the matter moves significantly slower than relativistic speeds and therefore contributes very little to the overall pressure. On the other hand, radiation has a momentum of the order of $\rho c$ and therefore has a non-negligible pressure.

$\textbf{3(b)}$ Integrate to find t for empty and flat universe models:

Let $\Omega_{r} = \Omega_{\Lambda} = 0$ for a pure matter universe.

$$t = \int_{0}^{1} \frac{da}{\dot{a}}$$ 

Use $\dot{a} = H(a) a$
$$\dot{a} = H_{0}[\frac{\Omega_{m}}{a} + (1 - \Omega_{m})]^{\frac{1}{2}}$$

$$t = \int_{0}^{1} \frac{da}{H_{0}[\frac{\Omega_{m}}{a} + (1 - \Omega_{m})]^{\frac{1}{2}}}$$ 

for $\Omega_{m} = 0$:

$$t = \int_{0}^{1} \frac{da}{H_{0}}$$

$$t = \frac{1}{H_{0}} = 14.4 Gyr$$

for $\Omega_{m} = 1$:

$$t = \int_{0}^{1} \frac{a^{\frac{1}{2}} da}{H_{0}}$$

$$t = \frac{2}{3 H_{0}} = 9.6 Gyr$$

$\textbf{3(c)}$ Find t for $\Omega_{m} = 0.27$:

$$t = \int_{0}^{1} \frac{da}{H_{0}(\frac{0.27}{a} + 0.73)^{\frac{1}{2}}}$$

$$t = 11.8 Gyr$$

This time is still too short, given that the oldest globular clusters have an age of $13 Gyr$. Using pure matter models, it is very difficult to get an age of the universe that matches observed data without using unrealistically small amounts of mass. While the empty universe model calculated in part (b) gives an old enough age, we of course know that it is not correct. Using the real value for $\Omega_{m}$ produces too short of a time, showing that some other factors must contribute.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import astropy.constants as const
import astropy.units as u
import scipy.integrate as integrate


#define constants:
H0 = 72.
c = const.c.to(u.km/u.s).value #use c in km/s to match units on hubble const
omegaM = 0.27
omegaL = 0.73
omegaR = 8.4e-5

count = 100

zpoints = np.linspace(0,5,count)

def get_dDcdz(z):
    x = 1./(1.+z)
    return c/H0 * 1/((x**(-4.)*omegaR + x**(-3.)*omegaM + x**(-2)*(1 - omegaM - omegaL) + omegaL)**(1./2.))

Dc = np.zeros(count)
Dl = np.zeros(count)
Da = np.zeros(count)
Hz = np.zeros(count)

for i in range(count):
    z = zpoints[i]
    x = 1./(1.+z)
    a = integrate.quad(get_dDcdz, 0, z)
    Dc[i] = a[0]
    Da[i] = Dc[i]/(1+z)
    Dl[i] = Dc[i]*(1+z)
    Hz[i] = H0 *(x**(-4.)*omegaR + x**(-3.)*omegaM + x**(-2)*(1 - omegaM - omegaL) + omegaL)**(1./2.)

    
fig = plt.figure(1, figsize=(10,40))
ax1 = fig.add_subplot(511)
ax1.plot(zpoints, Dc)
ax1.set_xlabel('Redshift (z)')
ax1.set_ylabel('Comoving Distance (Mpc)')
ax1.xaxis.grid(True)
ax1.yaxis.grid(True)

ax2 = fig.add_subplot(512)
ax2.plot(zpoints, Dl)
ax2.set_xlabel('Redshift (z)')
ax2.set_ylabel('Luminosity Distance (Mpc)')
ax2.xaxis.grid(True)
ax2.yaxis.grid(True)

ax3 = fig.add_subplot(513)
ax3.plot(zpoints, Da)
ax3.set_xlabel('Redshift (z)')
ax3.set_ylabel('Angular Distance (Mpc)')
ax3.xaxis.grid(True)
ax3.yaxis.grid(True)

ax4 = fig.add_subplot(514)
ax4.plot(zpoints, Hz)
ax4.set_xlabel('Redshift (z)')
ax4.set_ylabel('H(z)')
ax4.xaxis.grid(True)
ax4.yaxis.grid(True)

plt.show()    
    
                      
```
