```python
#4

c = 3.0E8 #m
H0 = 72. #km/s/Mpc convert t0 SI units
#H0 = 72 * 1000 * (1/(3.09E22))
Dh = c/H0

Om = 0.27
Ol = 0.73
Or = 8.4 * (10 ** -5)
```

```python
#z = np.linspace(0, 5, 100)

#E = np.sqrt(((1 + z)**4)*Or + ((1 + z)**3)*Om + ((1 + z)**2)*(1 - Om - Ol) + Ol)

#4a

def dcdz(z):
    dcdz0 = Dh * (1/(np.sqrt(((1 + z)**4)*Or + ((1 + z)**3)*Om + ((1 + z)**2)*(1 - Om - Ol) + Ol)))
    return dcdz0

#initialize M_b
Dc0 = 0
Dc = Dc0
Dc_val = list()

#create stepsize
r = np.linspace(0, 5, 100)

#calculate M_b
for i in r:
    Dc = Dc + dcdz(i) * (5/100.0)
    Dc_val.append(Dc)
    
fig = plt.figure(1, figsize=(8,8))

ax1 = fig.add_subplot(111)
ax1.plot(r, Dc_val)
ax1.set_xlabel("distance")
ax1.set_ylabel("co moving distance")

plt.show()
```
