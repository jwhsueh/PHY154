```python
import scipy

import numpy as np

import matplotlib.pyplot as plt

from scipy.integrate import odeint

from astropy import constants as constants

from astropy import units as units

%matplotlib inline

c=3.0E8/1000

H0=72

Omega_r=8.4E-5

Omega_m=0.27

Omega_y=0.73

D_H=c/H0


def E(z):
    
    Ez= np.sqrt(((1+z)**(4))*Omega_r + ((1+z)**(3))*Omega_m +((1+z)**(2))*(1-Omega_m-Omega_y)+Omega_y)

    return Ez


def Dc(z):
    
    dDc_dz= D_H*1/(E(z))
    
    return dDc_dz


def main(y, z): 
     
       
    derivs=[Dc(z)]


    return derivs
                       
                                   
                                   
y0=[0]
                                   
                                   
                                   
#redshift array for solution

z = np.linspace(0, 5, 1000)
                                   
#solution                                   
soln = odeint(main, y0, z)
                                   
#plotting

Dl=soln[:,0]*(1+z)

DA=soln[:,0]/(1+z)


Hz= E(z)*H0

  
fig = plt.figure(1, figsize=(10,12))




ax1 = fig.add_subplot(411)
ax1.plot(z, soln[:,0])
ax1.set_xlabel('Redshift (z) (Mpc) ')
ax1.set_ylabel('Comoving Distance')


ax2 = fig.add_subplot(412)
ax2.plot(z, Dl)
ax2.set_xlabel('Redshift (z) (Mpc) ')
ax2.set_ylabel('Angular Diameter Distance')

ax3 = fig.add_subplot(413)
ax3.plot(z, DA)
ax3.set_xlabel('Redshift (z) (Mpc) ')
ax3.set_ylabel('Luminosity Distance')
                                 
ax4 = fig.add_subplot(414)
ax4.plot(z, Hz)
ax4.set_xlabel('Redshift (z) (Mpc) ')
ax4.set_ylabel('Hubble Parameter')    


plt.tight_layout()
plt.show()
```
