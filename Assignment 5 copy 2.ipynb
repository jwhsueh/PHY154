```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Question 3

G = 6.67e-11
c = 3.0e8
radcon = 4.848 * (10 ** -6)

theta1 = 5.35 * radcon
theta2 = 0.8 * radcon
thetaE = ((theta1+theta2)/2)

zl = 0.36
zs = 1.41

Dl = 1047.3*3.086e22
Ds = 1768.2*3.086e22
Dls = Ds - Dl

b = thetaE*Dl

constE = (4*G)/(c*c*b)

#mass of the galaxy

M = (thetaE*Ds*((c)**2)*Dl)/(Dls*4*(G))
print "M =", M
```

```python
#Question 4 (a,b,c)

for i in range(1,40):
    for j in range(1,40):
        
        dx = i - 20.0
        dy = j - 20.0
        r = np.sqrt((dx*dx)+(dy*dy))
        
        if i != 20.0:
            bx = (dx*0.05*radcon)-(Dls*4.0*G*M)/(Ds*Dl*c*c*r*0.05*radcon)
        else:
            bx == 0.0
        if j != 20.0:
            by = (dy*0.05*radcon)-(Dls*4.0*G*M)/(Ds*Dl*c*c*r*0.05*radcon)
        else:
            by == 0.0
            
        R = ((bx*bx)+(by*by))**0.5
        if (4.74 < R < 5.0):
            plt.scatter(i,j)
```
