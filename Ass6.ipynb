```python
#4a) 
Hknot = 72.0 #km/s/Mpc
dH = c / Hknot

z = 0
omegaM = 0.27
omegaL = 0.73
omegaR = 8.4E-5

#Calculate
def dcdz(z):
    dcdz0 = dH * omegaR * (1 / (np.sqrt((1 + z)**4)) + omegaM * ((1 + z)**3)) + ((1 - omegaM - omegaL)*(1+z)**2 + omegaL)
    return dcdz0

dC = 0
dC_val = list()
r = np.linspace(0, 5, 100) #stepsize

for i in r:
    dC = dC + dcdz(i) * (5/100.0)
    dC_val.append(dC)
    
#plotting
fig = plt.figure(1, figsize=(8,8))
ax1 = fig.add_subplot(111)
ax1.plot(r, dC_val)
ax1.set_xlabel("Distance")
ax1.set_ylabel("Moving Distance")
```
