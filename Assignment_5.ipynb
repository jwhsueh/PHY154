```python
import numpy as np
import matplotlib.pyplot as plt

#constants
Ms = 2.0E30
G = 6.67 * (10 ** -11)
c = 3.0E8

#4b

#create array with n pixels, using Hubble parameters
n = 800 #number of pixels
#lens = np.zeros((n, n))
#source = np.zeros((n,n))

#info from previous problem
Ds = 1768.2 * 3.08E22 #meters
Dd = 1047.3 * 3.08E22 #meters
Dds = Ds - Dd

theta1 = 5.35 * 4.848 * (10 ** -6) #radians
theta2 = -0.8 * 4.848 * (10 ** -6) #radians
thetaE = (abs(theta1) + abs(theta2))/2.0

M = (Ds*Dd*c*c*thetaE)/(Dds*4*G)
```

```python
#4b/c
for i in range(1, 40):
    for j in range(1, 40):
        
        dx = i - 20
        dy = j - 20
        r1 = np.sqrt((dx*dx)+(dy*dy))
            
        if i != 20:
            bx = dx * 0.05 * (4.848 * (10 ** -6)) - (4.0*Dds*G*M)/(Ds*Dd*c*c*r1*(4.848 * (10 ** -6))*0.05)
        else:
            bx == 0
        if j != 20:
            by = dy * 0.05 * (4.848 * (10 ** -6)) - (4.0*Dds*G*M)/(Ds*Dd*c*c*r1*(4.848 * (10 ** -6))*0.05)
        else:
            by == 0
        #trig stuff
        r = np.sqrt(bx**2 + by**2)
        if ( r < 5):
            plt.subplot(211)
            plt.scatter(i, j)
            plt.subplot(212)
            plt.scatter(bx, by)

plt.show()
```
