```python
hubble1 = np.loadtxt('hubble1.dat', skiprows = 1,  usecols=range(1,7), unpack = True)
hubble2 = np.loadtxt('hubble2.dat', skiprows = 1, unpack = True)
## d: Mpc
## v: km/s
v1 = hubble1[2]
v2 = hubble2[5]
d1 = hubble1[1]
d2 = hubble2[4]
def func(x, k):
    return k*x
popt1, pcov1 = curve_fit(func, d1, v1)
popt2, pcov2 = curve_fit(func, d2, v2)
print popt1, popt2, pcov1,pcov2

## Hubble1
plt.figure()
plt.xlabel('Distance(Mpc)')
plt.ylabel('Velocity(km/s)')
plt.grid()
plt.xlim((-0.5,2.5))
plt.errorbar(d1, v1, linewidth=2, marker = "o", yerr = pcov1[0][0], fmt='o')
plt.plot(d1, d1 * popt1)
plt.title("$Hubble_1,\,H_0=423.9\,km\cdot s^{-1}\cdot Mpc^{-1}$",fontsize=20)


## Hubble2
plt.figure()
plt.xlabel('Distance(Mpc)')
plt.ylabel('Velocity(km/s)')
plt.grid()
plt.errorbar(d2, v2, linewidth=2, marker = "o", xerr = pcov2[0][0], fmt='o')
plt.plot(d2, d2 * popt2)
plt.title("$Hubble_2,\,H_0=71.4\,km\cdot s^{-1}\cdot Mpc^{-1}$",fontsize=20)
plt.show()
```

By Newtonian approach, in the homegenous sphere:
\begin{align}
    M &= \frac{4\pi}{3}r^3\rho = \frac{4\pi}{3}(ax)^3\rho\\
    ma &= -\frac{GMm}{r^2} = \ddot r m\\
    -\frac{4\pi}{3}G\rho &= \frac{\ddot r}{r} = \frac{\ddot{a} }{a}
\end{align}
That is:
\begin{align}
    \ddot{a} = -\frac{4\pi}{3}G\rho a
\end{align}
